```python
import pandas as pd
import altair as alt
```

```python
results_df = pd.read_csv('/content/drive/MyDrive/Machine Learning Deliverable/data/ML_results21_5.csv')
results_df
```

```python
values = ['VotingHard', 'Stacking', 'MLP-best', 'ExtrTrees-best2', 'RandomForest-best', 'SVM', 'DecisionTrees-best', 'LogisticRegression-best', 'GuassianNB-best', 'QDA', 'KNN', 'LDA']
```

```python
results_df = results_df.loc[results_df['Algorithm'].isin(values)]
results_df.reset_index(drop=True, inplace=True)
```

```python
results_df
```

```python
colorgorical = ["rgb(105,239,123)", "rgb(216,21,180)", "rgb(172,225,183)", "rgb(120,40,87)", "rgb(37,184,234)", "rgb(32,80,46)", "rgb(250,175,227)", "rgb(52,70,109)", "rgb(192,225,92)", "rgb(114,68,185)", "rgb(99,161,34)"]

chart = alt.Chart(results_df, width=alt.Step(25)).mark_bar().encode(
    alt.X('Algorithm:N', title='Algorithm', sort='-y'),
    alt.Y('F1_Avg:Q', title='Average F1'),
    alt.Color('Algorithm:N', title='Algorithm', scale = alt.Scale(scheme = colorgorical)),
)
chart
```
